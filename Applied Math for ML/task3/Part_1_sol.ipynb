# Problem 1

Use numpy to simulate the vector outcomes for repeating the following random experiments $1000$ times and use the relative frequency to compute the probability of the corresponding events:
    
1. **Experiment:** Flipping a fair coin, **Event:** A head occurs.
     
Observe how changing the number of repetition affects the computed probabilities.

**Hints:** 
- Use [numpy.random.randint](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html) or [numpy.random.choice](https://numpy.org/doc/stable/reference/random/generated/numpy.random.choice.html) to simulate the experiment.
- Other useful functions [numpy.arange](https://numpy.org/doc/stable/reference/generated/numpy.arange.html), [numpy.cumsum](https://numpy.org/doc/stable/reference/generated/numpy.cumsum.html), [numpy.sum](https://numpy.org/doc/stable/reference/generated/numpy.sum.html), and [matplotlib.pyplot.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
trails = [10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000]
probabilities1 = []
for n in trails:
    # 0 = tails, 1 = heads
    flips = np.random.randint(0, 2, n)
    num_heads = np.sum(flips)
    prob_heads = num_heads / n 
    probabilities1.append(prob_heads)
    print(f"n = {n}, P(e) = {prob_heads:.5f}")
plt.plot(trails,probabilities1 , label='Simulated Probability')   

```

2. **Experiment:** Rolling a fair die, **Event:** Number 5 occurs.

```python
trails = [10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000]
probabilities2 = []
for n in trails:
    #(1 and 6)
     rolls = np.random.randint(1, 7, n)
     count_fives = np.count_nonzero(rolls == 5)
     prob_five = count_fives / n 
     print(f"n = {n}, P(e) = {prob_five:.5f}")
     probabilities2.append(prob_five)
    
plt.plot(trails,probabilities2 , label='Simulated Probability')   
```

3. **Experiment:** Rolling a fair die, **Event:** A number greater than 3 occurs.

```python
trails = [10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000]
probabilities3 = []
for n in trails:
    #(1 and 6)
     rolls = np.random.randint(1, 7, n)
     count_greaterthan_three = np.count_nonzero(rolls > 3)
     prob_greaterthan_three = count_greaterthan_three / n 
     print(f"n = {n}, P(e) = {prob_greaterthan_three:.5f}")
     probabilities3.append(prob_greaterthan_three)
    
plt.plot(trails,probabilities3 , label='Simulated Probability')   
```

4. **Experiment:** Rolling two fair dice, **Event:** Sum of two dice equal 5.

```python
trails = [10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000]
probabilities4 = []
for n in trails:
    #(1 and 6)
     rolls1 = np.random.randint(1, 7, n)
     rolls2 = np.random.randint(1, 7, n)
     count_sum_equal_5 = np.count_nonzero(rolls1 + rolls2 == 5)
     prob_sum_equal_5 = count_sum_equal_5 / n
     probabilities4.append(prob_sum_equal_5)
     print(f"n = {n}, P(e) = {prob_sum_equal_5:.5f}")
     probabilities3.append(prob_sum_equal_5)
    
plt.plot(trails,probabilities4 , label='Simulated Probability')   

```

# Problem 2

Simulate the experiment of tossing $1000$ times a coin with probability of **head (coded as $1$) equal to $0.4$** and probability of **tail (coded as $0$) equal to $0.6$**. How can you check if the simulation of the experiment of been running as expected.

**Hint**: Use [matplotlib.pyplot.hist¶
](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)

```python
tosses = np.random.choice([1, 0], size=1000, p=[0.4, 0.6])#غريبه عليا
count_heads = np.sum(tosses)  # نعد الوحايد
count_tails = 1000 - count_heads  
prob_heads = count_heads / 1000
prob_tails = count_tails / 1000
print(f"Number of heads: {count_heads}, Probability of heads: {prob_heads:.2f}")
print(f"Number of tails: {count_tails}, Probability of tails: {prob_tails:.2f}")

# Plot histogram دي جبتها من شات عشان ابين الشكل
plt.hist(tosses, bins=[-0.5, 0.5, 1.5], edgecolor='black', rwidth=0.8)
plt.xticks([0, 1], labels=["Tails (0)", "Heads (1)"])
plt.title("Histogram of Coin Toss Outcomes")
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.show()
```

# Problem 3: Independence

Use NumPy to create simulations and compute **empirical  probability** for the following outcomes. Compare the results with the **theoretical probability**.

1. Two fair coin flips produce exactly two heads

```python

# simulate 1 million tests of two fair coin flips
flip1 = np.random.randint(0, 2, 1000000)  # First coin flip
flip2 = np.random.randint(0, 2, 1000000)  # Second coin flip

# sums of all tests
sums = flip1 + flip2 
# proportion of tests that produced exactly two heads
two_heads = np.sum(sums == 2)
proportion_two_heads = two_heads / 1000000

print(f" proportion of tests that produced exactly two heads : {proportion_two_heads:.5f}")
```

2. Three fair coin flips produce exactly one head

```python
#type your code here:

# simulate 1 million tests of three fair coin flips
flip1 = np.random.randint(0, 2, 1000000)  # First coin flip
flip2 = np.random.randint(0, 2, 1000000)  # Second coin flip
flip3 = np.random.randint(0, 2, 1000000)  # third coin flip

# sums of all tests
sums = flip1 + flip2 + flip3 

# proportion of tests that produced exactly one head
exactly_one_head = np.sum(sums == 1)
proportion_one_head = exactly_one_head / 1000000

print(f"Proportion of tests with exactly one head: {proportion_one_head:.5f}")
```

3. Three biased coin flips with P(H) = 0.6 produce exactly one head

```python
 
# simulate 1 million tests of three biased coin flips
# hint: use np.random.choice()
flip1 = np.random.choice([0, 1], size=1000000, p=[0.4, 0.6])  # First coin flip
flip2 = np.random.choice([0, 1], size=1000000, p=[0.4, 0.6])  # Second coin flip
flip3 = np.random.choice([0, 1], size=1000000, p=[0.4, 0.6])  # Third coin flip

# sums of all tests
sums = flip1 + flip2 + flip3 

# proportion of tests that produced exactly one head
exactly_one_head = np.sum(sums == 1)
proportion_one_head = exactly_one_head / 1000000

print(f"Proportion of trials with exactly one head: {proportion_one_head:.5f}")
```

# Problem 4:Random Variables Statistical Tools
Simulate Rolling a fair die $1000$ times. Then calculate the following:

- Sample mean
- Sample variance
- Sample standard deviation

Compare the results with the theoretical probability.

```python
rolls = np.random.randint(1, 7, 1000)
sample_mean = np.mean(rolls)
sample_variance = np.var(rolls)
sample_std_dev = np.std(rolls)

print(f"Sample mean: {sample_mean:.5f}")
print(f"Sample variance: {sample_variance:.5f}")
print(f"Sample standard deviation: {sample_std_dev:.5f}")

# Theoretical values
print("\nTheoretical values:")
print(f"Theoretical mean: 3.463")
print(f"Theoretical variance: 3.005636636636637")
print(f"Theoretical standard deviation: 1.7336772008181445")

```

# Problem 5: Correlation and Covariance

The CSV file includes a hourly/daily summary for Szeged, Hungary area, between 2006 and 2016. After inspecting the data:

- Calculate the covariance and correlation coeffecient between temperature and apparent temperature
- Calculate the covariance and correlation coeffecient between temperature and humidity

```python
df = pd.read_csv('weatherHistory.csv')
temperature = df['Temperature (C)'].to_numpy().reshape(-1, 1)
apparant_temp = df['Apparent Temperature (C)'].to_numpy().reshape(-1, 1)
humidity = df['Humidity'].to_numpy().reshape(-1, 1)
```

```python
plt.scatter(temperature, apparant_temp)
plt.grid()
plt.xlabel('Temperature (C)')
plt.ylabel('Apparent Temperature (C)')
```

```python
plt.scatter(humidity, temperature)
plt.grid()
plt.xlabel('Humidity')
plt.ylabel('Temperature (C)')
```

## Temperature and Apparent Temperature

```python
cov_temp_apparent = np.cov(temperature, apparant_temp)[0,1]
corr_temp_apparent = np.corrcoef(temperature, apparant_temp)[0,1]
 
print(f"Covariance between Temperature and Apparent Temperature: {cov_temp_apparent:.5f}")
print(f"Correlation coefficient between Temperature and Apparent Temperature: {corr_temp_apparent:.5f}")

 
```

## Temperature and Humidity

```python
cov_temp_humidity = np.cov(temperature, humidity) 
corr_temp_humidity = np.corrcoef(temperature, humidity)


print(f"Covariance between Temperature and Humidity: {cov_temp_humidity:.5f}")
print(f"Correlation coefficient between Temperature and Humidity: {corr_temp_humidity:.5f}")
```
