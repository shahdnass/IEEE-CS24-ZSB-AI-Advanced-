# Problem 1
a) Use the numpy function [numpy.random.normal](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html) to generate a standard Gaussian random variable $X$ (zero mean and unit variance).

- Use histogram to plot the estimated PDF of the random variable $X$
- Compute the true "analytical" PDF -using the equation- of $X$ . Plot the true PDF on the top of the histogram figure.
- Calculate the sample mean and the sample variance.

### True/Analytical PDF


## using `numpy`

```python

data = np.random.normal(loc=0, scale=1, size=10000) 
mu=sample_mean = np.mean(data)
sample_variance = np.var(data)
sigma = np.sqrt(sample_variance) 
plt.hist(data, bins=50, density=True, alpha=0.6, color='g', label="Estimated PDF")
x = np.linspace(-4, 4, 1000) 
true_pdf = (1 / (np.sqrt(2 * np.pi * sigma**2))) * np.exp(-((x - mu)**2) / (2 * sigma**2))
plt.plot(x, true_pdf, 'r', label="True PDF")
plt.title("Estimated vs True PDF of a Standard Gaussian")
plt.xlabel("X")
plt.ylabel("Density")
plt.legend()
print(f"Sample Mean: {sample_mean:.5f}")
print(f"Sample Variance: {sample_variance:.5f}")
plt.show()
```

## using `scipy`

```python
from scipy.stats import norm 
data = np.random.normal(loc=0, scale=1, size=10000)
mu = np.mean(data)
sample_variance = np.var(data)
sigma = np.sqrt(sample_variance)

plt.hist(data, bins=50, density=True, alpha=0.6, color='g', label="Estimated PDF")
x = np.linspace(-4, 4, 1000)
true_pdf = norm.pdf(x, loc=mu, scale=sigma)
plt.plot(x, true_pdf, 'r', label="True PDF")

plt.title("Estimated vs True PDF of a Standard Gaussian")
plt.xlabel("X")
plt.ylabel("Density")
plt.legend()

print(f"Sample Mean: {mu:.5f}")
print(f"Sample Variance: {sample_variance:.5f}")

plt.show()
```

- Compute the true "analytical" PDF using [scipy.stats.norm.pdf](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html). Plot the true PDF on the top of the histogram figure.
- Compute the CDF using [scipy.stats.norm.cdf](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), then plot it.

```python
 X = np.random.normal(0, 1, 10000)

sample_mean = np.mean(X)
sample_variance = np.var(X)

x_values = np.linspace(-4, 4, 1000)

plt.hist(X, bins=50, density=True, alpha=0.6, color='g')
plt.plot(x_values, stats.norm.pdf(x_values, loc=0, scale=1), 'b')
plt.plot(x_values, stats.norm.cdf(x_values, loc=0, scale=1), 'r')

plt.grid(True)
plt.show()
```

b) Repeat (a) for a Gaussian random variable $Y$ with $mean=3$ and $variance=4$. 

```python
mu = 3  
sigma = 2 
data = np.random.normal(loc=mu, scale=sigma, size=10000)
sample_mean = np.mean(data)
sample_variance = np.var(data)
plt.hist(data, bins=50, density=True, alpha=0.6, color='g', label="Estimated PDF")
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)  
true_pdf = norm.pdf(x, loc=mu, scale=sigma)  
plt.plot(x, true_pdf, 'r', label="True PDF")

plt.title("Estimated vs True PDF of a Gaussian (Mean=3, Variance=4)")
plt.xlabel("Y")
plt.ylabel("Density")
plt.legend()

print(f"Sample Mean: {sample_mean:.5f}")
print(f"Sample Variance: {sample_variance:.5f}")


plt.show()
```

```python
mean = 3  # Mean
variance = 4  # Variance
std_dev = np.sqrt(variance)  
Y = np.random.normal(loc=mean, scale=std_dev, size=10000)
sample_mean = np.mean(Y)
sample_variance = np.var(Y)
x_values = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
plt.hist(Y, bins=50, density=True, alpha=0.6, color='g', label="Estimated PDF")

plt.plot(x_values, stats.norm.pdf(x_values, loc=mean, scale=std_dev), 'b', label="True PDF")
plt.plot(x_values, stats.norm.cdf(x_values, loc=mean, scale=std_dev), 'r', label="True CDF")
plt.title("Estimated vs True PDF and CDF of a Gaussian (Mean=3, Variance=4)")
plt.xlabel("Y")
plt.ylabel("Density")
plt.legend()

# Add grid
plt.grid(True)

# Display the plot
plt.show()

# Print the sample mean and sample variance
print(f"Sample Mean: {sample_mean:.5f}")
print(f"Sample Variance: {sample_variance:.5f}")
```

e) Repeat for an binomial random variable with $n = 5$ and $p = 0.5$

```python
#type your code here:
from scipy.stats import binom
n = 5
p = 0.5
sample = np.random.binomial(n, p, 1000)
sample_mean = np.mean(sample)
sample_variance = np.var(sample)

print(f"Sample mean: {sample_mean}")
print(f"Sample variance: {sample_variance}")
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
plt.figure(figsize=(10, 5))
plt.bar(x, pmf, color='blue')
plt.title('PMF of Binomial Random Variable')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
cdf = binom.cdf(x, n, p)
plt.figure(figsize=(10, 5))
plt.step(x, cdf, where='post', color='red')
plt.title('CDF of Binomial Random Variable')
plt.xlabel('Number of Successes')
plt.ylabel('Cumulative Probability')
plt.show()
```

# Problem 3: Practicing the Central Limit Theorem
Use Python to simulate the Central Limit Theorem for the following random i.i.d variables and check that the higher the number of number the variables, the more the mean of subsets behave like a gaussian random variable.


```python
sample_sizes = [2, 50]  
means_2 = [np.mean(np.random.uniform(0, 1, 2)) for _ in range(10000)]
means_50 = [np.mean(np.random.uniform(0, 1, 50)) for _ in range(10000)]

plt.figure(figsize=(10, 5))
plt.hist(means_2, bins=30, alpha=0.6,  color='red', density=True)
plt.hist(means_50, bins=30, alpha=0.6, color='blue', density=True)

plt.title("Central Limit Theorem ")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
